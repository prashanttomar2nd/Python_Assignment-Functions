10. 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
    - Map : The map function's job is to transform every item in an iterable. It applies a given function to each element and returns an iterator with the results of that transformation.

      reduce : The reduce function's job is to aggregate a sequence of values into a single, cumulative result. It works by repeatedly applying a function to the items of the iterable, from left to right, reducing the iterable to a single value.

      filter : The filter function's job is to select items from an iterable. It applies a function that returns True or False to each element. It then returns an iterator containing only the items for which the function returned True.

```python
#  1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def even_number_all(numbers):
  return sum(num for num in numbers if num % 2 == 0)

even_number_all([1,2,3,4,5,6,7,8])
```

```python
# 4.  Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime_number(n):
  if n <= 1:
    return False

  for i in range(2, n):
    if n % 2 == 0:
      return False

  return True

prime_number(19)
```

```python
# 5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fib(n):
  fib_seq = []
  a = 0
  b = 1
  for i in range(n):
    a,  b = b, a + b
    fib_seq.append(a)
  return fib_seq
fib(10)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def exponent_power(n):
  for i in range(n+1):
    yield 2 ** i

exponent_power(4)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
temp_list = [42,56,23,63]

Fahrenheit_list = list(map(lambda x : x * 9/5 + 32, temp_list))

Fahrenheit_list
```

```python
# 10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

vowels = "aeiouAEIOU"

def remove_all_vowels(s):
  return tuple(filter(lambda x : x not in vowels, s))

remove_all_vowels("PrashantOrange")
```
